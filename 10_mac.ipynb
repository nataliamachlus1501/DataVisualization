```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
```

```python
#pobieranie danych
import urllib.request 
url = "https://raw.githubusercontent.com/dwyl/english-words/master/words_alpha.txt"
file = urllib.request.urlopen(url)
slowa1 = []
for line in file:
    linia = line.decode("utf-8")
    slowa1.append(linia)
```

```python
slowa1
```

```python
#zamiana wszystkich słow na słowa o małych literach
for i in range(len(slowa1)):
    slowa1[i]=slowa1[i].lower()
```

```python
slowa1
```

```python
#formatowanie listy
slowa = []
for slowo in slowa1:
    slowa.append(slowo.strip())
```

```python
slowa
```

```python
#lista wszystkich dwóch pierwszych liter słow
lista_par=[]
for i in range(len(slowa)):
    if len(slowa[i]) >= 2:
        lista_par.append(slowa[i][:2])
```

```python
lista_par
```

```python
from collections import Counter
from itertools import combinations
zliczacz = Counter(lista_par)
zliczacz
```

```python
zliczacz['aa']
```

```python
zliczacz['un']
```

```python
type('nu')
```

```python
most_common = list(zliczacz.most_common())
most_common
```

```python
paryl=[]
for i in range(0,len(most_common)):
    paryl.append(most_common[i][0])
paryl
```

```python
czestosc=[]
for i in range(0,len(most_common)):
    czestosc.append(most_common[i][1])
czestosc
```

```python
#convert it to dictionary with values and its occurences
from wordcloud import WordCloud
from collections import Counter
word_cloud_dict = (" ".join(lista_par))
wordcloud = WordCloud(width = 1000, height = 1000, collocations = False,random_state = 1, colormap = "YlGnBu").generate(word_cloud_dict)
plt.figure(figsize=(10,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
matrix=[]
num_letters=26
letters=["abcdefghijklmnopqrstuvwxyz"]
tupple=""
for i in range(num_letters):
    matrix.append([])
    for j in range(num_letters):
        para = letters[0][i]+letters[0][j]
        matrix[i].append(zliczacz[para])
```

```python
matrix
```
