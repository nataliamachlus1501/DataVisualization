```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sympy as sym
from sympy import *
```

```python
T0=[0.2408467,0.61519726,1.0000174,1.8808158,11.862615,29.447498,84.016846,164.79132]
r0=[0.3870993,0.723336,1.000003,1.52371,5.2029,9.537,19.189,30.0699]
```

```python
r=[]
for i in range(8):
    r.append((r0[i])*1.49597870700*10**11)
r
```

```python
T=[]
for i in range(8):
    T.append((T0[i])*31557600)
T
```

```python
T2=[]
for i in range(8):
    T2.append((T[i])*T[i])
T2
```

```python
r3=[]
for i in range(8):
    r3.append(((r[i])*(r[i])*r[i]))
r3
```

```python
plt.style.use('dark_background')
fig=plt.figure(figsize=(10,6))
ax=fig.add_subplot()
ax.plot(T2,r3, color='aqua', linewidth='2')
ax.set_xlabel('Okres orbitalny(T)', size = 10) 
ax.set_ylabel('Półoś wielka(r)', size = 10)

ax.set_yscale('log')
ax.set_xscale('log')
ax.text(10**14,10**37,r"$\frac{ T^{2} }{ r^{3} }$=const", size=20)

ax.scatter(T2[0],r3[0], color = 'grey',label = 'merkury',s = 10, zorder = 2.5)
ax.scatter(T2[1],r3[1], color = 'pink', label = 'wenus',s = 40, zorder = 2.5)
ax.scatter(T2[2],r3[2], color = 'green',label = 'ziemia', s = 50, zorder = 2.5)
ax.scatter(T2[3],r3[3], color = 'yellow', label = 'mars',s = 30, zorder = 2.5)
ax.scatter(T2[4],r3[4], color = 'red', label = 'jowisz',s = 100, zorder = 2.5)
ax.scatter(T2[5],r3[5], color = 'brown',label = 'saturn', s = 80, zorder = 2.5)
ax.scatter(T2[6],r3[6], color = 'lightblue',label = 'uran',s = 70, zorder = 2.5)
ax.scatter(T2[7],r3[7], color = 'blue',label = 'neptun', s = 60, zorder = 2.5)
ax.legend(loc = 'lower right')
plt.title("Trzecie prawo Keplera", size = 13)
plt.savefig('03_mac_273344_kropki.pdf')
plt.show()
```
