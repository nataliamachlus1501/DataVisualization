```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sympy as sym
from sympy import *
import random
```

```python
przedzialy= random.randint(0,30)
przedzialy
```

```python
from scipy.integrate import quad

def integrand(x):
    return 2/(2+x**2)

ans, err = quad(integrand, 0, 5)
pole = ans
ans
```

```python
n=przedzialy
x = np.linspace(0,5,n+1)
y = 2/(2+x**2)
X=np.linspace(0,5,n*(n+1))
Y=2/(2+X**2)

fig, (ax1,ax2,ax3) = plt.subplots(nrows = 1, ncols = 3, sharex = True, figsize = (12,6))

ax1.plot(X,Y)
ax2.plot(X,Y)
ax3.plot(X,Y)


x1_left = x[:-1] 
y1_left = y[:-1]

ax1.plot(x1_left,y1_left,'b.',markersize = 10)
ax1.bar(x1_left,y1_left,width=5/n,alpha=0.2,align='edge',edgecolor='b')

x3_right = x[1:] 
y3_right = y[1:]

ax3.plot(x3_right,y3_right,'b.',markersize = 10)
ax3.bar(x3_right,y3_right,width=-5/n,alpha=0.2,align='edge',edgecolor='b')
ax2.fill_between(X,Y, facecolor='lightpink', alpha=0.5)

#do lewego
pole_l = 0
for i in range(0,n):
    pole_l = pole_l +y1_left[i]*(5/n)
#do prawego
pole_r = 0
for i in range(0,n):
    pole_r = pole_r +y3_right[i]*(5/n)
    
ax1.set_ylim(bottom=0)
ax2.set_ylim(bottom=0)
ax3.set_ylim(bottom=0)

ax1.set_xlim(left=0)
ax2.set_xlim(left=0)
ax3.set_xlim(left=0)

ax2.text(2,0.8,r"$f(x)=\frac{ 2 }{ (3+x^{2})}$", size=16)
ax1.set_title("suma I =" +str(round(pole_l,5)))
ax3.set_title("suma II =" +str(round(pole_r,5)))
fig.suptitle('Przedziałów jest' +' '+ str(przedzialy)+','+ ' '+'pole ≈ '+' '+str(pole))
plt.savefig("05_mac_273344.pdf")
```
