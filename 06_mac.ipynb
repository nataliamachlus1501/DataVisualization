```python
import numpy as np
import matplotlib.pyplot as plt
import urllib.request 
```

```python
#pobieranie danych
url = "http://www.mbnet.com.pl/dl.txt"
file = urllib.request.urlopen(url)
wyniki = []
for line in file:
    linia = line.decode("utf-8")
    wyniki.append(linia)
```

```python
len(wyniki)
```

```python
liczby1 =[x.split()[2] for x in wyniki]
```

```python
#6000 ostatnich wyników
liczby=liczby1[-6000:]
liczby
```

```python
len(liczby)
```

```python
liczby = list(map(lambda x: x.split(','), liczby))
```

```python
from collections import Counter
from itertools import combinations
zliczacz=Counter()
for wynik in liczby:
    wynik.sort()
    for para in combinations(wynik,2):
        zliczacz[para] += 1
rezultat=zliczacz.most_common(20)
rezultat
```

```python
#lista przechowująca pary
lista1=[]
for i in range(0,20):
    lista1.append(rezultat[i][0])  
lista1
```

```python
#lista przechowująca częstość par
lista2=[]
for i in range(0,20):
    lista2.append(rezultat[i][1])  
lista2
```

```python
#funkcja licząca ilość par parzystych
liczba_wynikow = len(liczby)
for i in range(liczba_wynikow):
    for j in range(6):
        liczby[i][j]=int(liczby[i][j])
liczba_parzystych=[]
for i in range(7):
    liczba_parzystych.append(0)

for wynik in liczby:
    parzyste=0
    for i in range(6):
        if wynik[i] % 2 == 0:
            parzyste+=1
    liczba_parzystych[parzyste]+=1
print(liczba_parzystych)
#indeks 0 : parzystych 0, niep 6; indeks 1: parz 1 niep 5...
```

```python
#zamiana na string
lista_par = list(map(str,lista1))
lista_par
```

```python
#funkcja tych samych kolorów dla równych częstości
import  seaborn as sns
kolory = sns.color_palette("vlag",len(lista2))
for i in range(1,20):
    if lista2[i]==lista2[i-1]:
        kolory[i]=kolory[i-1]
```

```python
#zmienne potrzebne do wykresu
data = [x.split()[1] for x in wyniki]
p_los = data[0]
ost_los = data[-1]
l_los = len(wyniki)
```

```python
def Average(lst):
    return sum(lst) / len(lst)
```

```python
#wartość lini referencyjnej
Average(list(zliczacz.values()))
```
