```python
f = open("pan-tadeusz.txt",'r',encoding = 'utf-8')
text = f.read().split()
f.close()
```

```python
text[100]
```

```python
slowa = list(text)
```

```python
slowa[100:110]
```

```python
slowa2 = list(map(lambda x: x.replace(',',''), slowa))
slowa3 = list(map(lambda x: x.replace('.',''), slowa2))
slowa4 = list(map(lambda x: x.replace('?',''), slowa3))
slowa5 = list(map(lambda x: x.replace('!',''), slowa4))
slowa6 = list(map(lambda x: x.replace(':',''), slowa5))
slowa7 = list(map(lambda x: x.replace(';',''), slowa6))
```

```python
slowa2=list(map(lambda x: x.lower(), slowa2))
slowa3=list(map(lambda x: x.lower(), slowa3))
slowa4=list(map(lambda x: x.lower(), slowa4))
slowa5=list(map(lambda x: x.lower(), slowa5))
slowa6=list(map(lambda x: x.lower(), slowa6))
slowa7=list(map(lambda x: x.lower(), slowa7))
```

```python
dł = []
for i in slowa7:
    if len(i) >= 5:
        dł.append(i)
```

```python
dł
```

```python
len(dł)
```

```python
from collections import Counter
```

```python
counter = Counter(dł)
counter
```

```python
len(counter)
```

```python
mostcommon = list(counter.most_common(20))
mostcommon
```

```python
lista1=[]
for i in range(0,20):
    lista1.append(mostcommon[i][0])  
lista1
```

```python
lista2=[]
for i in range(0,20):
    lista2.append(mostcommon[i][1])
lista2
```

```python
import matplotlib.pyplot as plt
import  seaborn as sns
```

```python
plt.style.use('default')
plt.figure(figsize = (12,8))
fig,ax=plt.subplots()
plt.barh(lista1,lista2,align = 'center',color = sns.color_palette("vlag",len(lista1)))

plt.title('Adam Mickiewicz "Pan Tadeusz" \n 20 najczęściej wystepujących słów o długości conajmniej 5 liter:')

for i,v in enumerate(lista2):
    ax.text(v+3,i+0.3,lista1[i], color ='black')
    
for i,v in enumerate(lista2):
    ax.text(v/2,i+0.3,str(v), color = 'black')
    
ax.spines['top'].set_color('none')    
ax.spines['right'].set_color('none')
ax.set_yticklabels("",minor = False)
ax.invert_yaxis()
plt.savefig('04_mac_273344.svg')
plt.show()
```
